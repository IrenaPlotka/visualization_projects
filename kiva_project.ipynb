### Visualizing Loans Awarded by Kiva

In this project I'll visualize insights using a dataset from Kaggle. The dataset contains information about loans awarded by the non-profit Kiva.

Using Seaborn, I'll explore the average loan amount by country using aggregated bar charts. I'll also visualize the distribution of loan amounts by project type and gender using box plots and violin plots.

```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns

#importing the csv file into DataFrame 'df'
df = pd.read_csv(r'C:\Users\iplot\Downloads\seabornProjectCumulative\seabornProjectCumulative\kiva_data.csv')
#examining the data
print(df.head(20))
```

```python
# first adding Matplotlib Ticker to display loan_amount ticks on the y-axis begin with a $
import matplotlib.ticker as mtick

# Creating a bar plot to visualize the average size of Kiva loans given to projects, by country, and "nesting" by gender.
sns.set_palette("Set1")
sns.set_style('white')
sns.set_context('talk')
f, ax = plt.subplots(figsize=(14, 10))
sns.barplot(data=df, x='country', y='loan_amount', hue='gender')
fmt = '${x:,.0f}'
tick = mtick.StrMethodFormatter(fmt)
ax.yaxis.set_major_formatter(tick)
plt.title('Average Loan Amounts Given by Kiva, by country, by gender (in $ USD)')
sns.set_palette("Accent")
sns.set_style("ticks")
```

From the bar plot we can conclude that, in general, male projects get higher loans in all the countries; El Salvator has the least disparity in loan amounts awarded by gender.

```python
# Creating a box plot to visualize the distribution of Kiva loans given to projects, by country.
plt.figure(figsize=(14, 10))
sns.set_palette("Set2")
sns.boxplot(data=df, x="country", y="loan_amount")
plt.title('Distribution of Loan Amounts Given by Kiva, by country')
```

From the box plot we can conclude that Kenya has the biggest box distribution and we are likely to get a higher loan in Cambodia.

```python
# visualizing loan amount by activity
plt.figure(figsize=(14, 10))
sns.set_palette("Accent")
sns.set_style("ticks")
sns.boxplot(data=df, x="activity", y="loan_amount")
plt.title('Distribution of Loan Amounts Given by Kiva, by activity')
```

This visualization reveals that farming projects get higher loans. We can also have a look at the distribution of loans awarded by type of activity using violin plot.

```python
plt.figure(figsize=(16, 10))
sns.set_style("whitegrid")
sns.violinplot(data=df, x="activity", y="loan_amount")
```

From the violin plots one can conclude that Food Production/Sales and General Store activity projects get loans of more than $600 very rarely, while loans for Farming purposes are more evenly distributed and can reach up to $600 - $1000 more often.

```python
# Creating a violin plot that visualizes the distribution of loan amount by country and gender
sns.set_style('whitegrid')
sns.set_palette("Spectral")
plt.figure(figsize=(18, 12))
sns.violinplot(data=df, x="country", y="loan_amount", hue="gender", split=True)
plt.title('Distribution of Loan Amounts Given by Kiva, by country, by activity')
```

In Pakistan, the average loan amount received by women is much less than the average loan amount received by men.
The same can be said for Cambodia (more or less).
In El Salvador, there is a fairly equal distribution of loans (total number of loans) between genders, as well as
similar median loan amounts (on average).
